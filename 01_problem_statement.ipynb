```python
import math
import numpy as np
from numpy import random
import pandas as pd

import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
def read_points(point_count):
    return pd.read_csv("data/points_" + str(point_count).zfill(4) + ".csv", dtype={'label': str})
```

```python
# Functions to plot points and labels

def plot_result(img, points):
    dpi = 100
    margin = 0.01
    w = (1+2*margin) * img.shape[1] / float(dpi)
    h = (1+2*margin) * img.shape[0] / float(dpi)
    fig = plt.figure(figsize=(w, h), dpi=dpi)
    ax = fig.add_axes([margin, margin, 1/(1+2*margin), 1/(1+2*margin)])
    ax.axis('off')

    # Plot point marks
    ax.scatter(points.x, points.y, s=40, marker="o", cmap="tab20")
    
    # Plot independent point labels
    for i in range(len(points)):
        ax.text(points.x[i], points.y[i]-13, points.label[i], fontsize=9,
                ha='center', va='baseline',
                bbox=dict(boxstyle='round,pad=0.2', facecolor='white', alpha=0.8))
        
    # Plot the floor plan
    ax.imshow(img)
    plt.savefig("output/problem_" + str(len(points)) + ".png", interpolation='none')
    plt.show()
```

```python
img = mpimg.imread("floorplan.png")
points = read_points(50)
plot_result(img, points)
```
